```
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
iris = sns.load_dataset("iris")

```

```
print(iris)
```

```
iris[iris.columns[2:4]]
```

```
sns.kdeplot(iris[iris.columns[2:4]])
```

```
df = iris
setosa = df.loc[df['species'] == "setosa"]
virginica = df.loc[df['species'] == "virginica"]
```

```
plt.subplot(121)

sns.lineplot(x="petal_length", y="petal_width",
             data=setosa[setosa.columns[2:4]])
plt.title("setosa")
plt.subplot(122)
sns.lineplot(x="petal_length", y="petal_width",
             data=virginica[virginica.columns[2:4]])
plt.title("virginica")
```
